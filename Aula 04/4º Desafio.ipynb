```python
import pandas as pd

ibge = 'https://gist.githubusercontent.com/tgcsantos/85f8c7b0a2edbc3e27fcad619b37d886/raw/a4954781e6bca9cb804062a3eea0b3b84679daf4/Basico_SP1.csv'
#separa por ;
dados_ibge = pd.read_csv(ibge, sep=";",
                        encoding='ISO-8859-1',
                        decimal=',',
                        thousands='.')
dados_ibge.dropna(how='all', axis=1, inplace=True)


dados_ibge.drop(['Cod_setor', 
                 'Cod_Grandes Regiï¿½es',
                 'Nome_Grande_Regiao',
                 'Cod_UF',
                 'Nome_da_UF ',
                 'Nome_da_micro',
                 'Cod_meso',
                 'Nome_da_meso',
                 'Cod_micro', 
                 'Cod_RM',
                 'Nome_da_RM',
                 'Cod_municipio',
                 'Nome_do_municipio',
                 'Cod_distrito',
                 'Cod_subdistrito', 
                 'Cod_bairro', 
                 'Nome_do_bairro',
                 'Situacao_setor',
                 'V001',
                 'V002',
                 'V004'],axis=1, inplace=True)

dados_ibge.replace('ï¿½', 'A', regex=True, inplace=True)

dados_ibge.head()
```

```python
data = url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(data)

dados[['primeira', 'Novo_valor', 'ultima']] = dados['Valor'].str.split(expand=True)
dados = dados.drop(['primeira', 'ultima'], axis=1)
dados['Novo_valor'] = dados['Novo_valor'].str.replace('.','').astype(float)
dados.head()
```

## Relação entre quantidade de pessoas x valor nominal de renda

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10,10))
ax = sns.scatterplot(data=dados_ibge, x='V005', y='V003')
ax.set_title('Quantidade média moradores X Valor médio mensal de renda')
ax.set_xlabel('Valor do redimento nominal médio', labelpad=25)
ax.set_ylabel('Média de pessoas por domicílio', labelpad=25)
ax.legend(labels=['População X Valores'])
plt.show()
```

## Verificando correlação entre colunas - com gráficos

```python
plt.figure(figsize=(15,6))
mask = np.triu(np.ones_like(dados_ibge.corr(), dtype=bool))
hm=sns.heatmap(dados_ibge.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
hm.set_title('Correlação', fontdict={'fontsize':18}, pad=16)
plt.show()
```
