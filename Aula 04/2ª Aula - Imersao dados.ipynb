### Base ibge censo 2010 dos imoveis sp - importing

```python
import pandas as pd

ibge = 'https://gist.githubusercontent.com/tgcsantos/85f8c7b0a2edbc3e27fcad619b37d886/raw/a4954781e6bca9cb804062a3eea0b3b84679daf4/Basico_SP1.csv'
#separa por ;
dados_ibge = pd.read_csv(ibge, sep=";",
                        encoding='ISO-8859-1',
                        decimal=',',
                        thousands='.')
dados_ibge.dropna(how='all', axis=1, inplace=True)
dados_ibge.head()
```

```python
dados_ibge.info()
```

### Importando base formatada da aula 03 - base imoveis_sp

```python
data = url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(data)

dados[['primeira', 'Novo_valor', 'ultima']] = dados['Valor'].str.split(expand=True)
dados = dados.drop(['primeira', 'ultima'], axis=1)
dados['Novo_valor'] = dados['Novo_valor'].str.replace('.','').astype(float)
dados.head()
```

### Importando base de endereços e ceps/ lati e longe ibge

```python
enderecos = pd.read_csv('/home/keven/Documents/Dados/enderecos.csv')
enderecos
```

```python
enderecos_sp = enderecos.query("sigla_uf == 'SP'")
enderecos_sp
```

```python
enderecos_sp['rua'] = enderecos_sp['tipo_logr'] +' '+ enderecos_sp['logr_nome']
enderecos_sp['rua'] = enderecos_sp['rua'].str.lower().str.strip()
enderecos_sp.head()
```

```python
dados["Rua"].str.extract(r'(^[\w ]+)')[:10]
```

```python
dados["Rua"].str.extract(r'(^[\w ]+)')
```

```python
dados["apenas_rua"] = dados["Rua"].str.extract(r'(^[\w ]+)')
dados["apenas_rua"] = dados["apenas_rua"].str.lower().str.strip()
dados.head()
```

### Cruzando bases de dados endereços com dados, juntando uma só

```python
dados_geo = pd.merge(left=dados, 
                     right=enderecos_sp[['rua', 'cep', 'latitude','longitude']],
                     how='left',
                     left_on='apenas_rua',
                     right_on='rua').drop_duplicates(subset=dados.columns).query('cep > 0')
dados_geo
```
